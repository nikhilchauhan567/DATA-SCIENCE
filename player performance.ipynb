```python
# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

#drop rows with NaN values after replacing inf values
df.dropna(inplace=True)
```

```python
# Function to remove outliers using IQR method
def remove_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    # Filter out outliers
    df_cleaned = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_cleaned

# Apply the function to remove outliers from all numerical columns
data_cleaned = remove_outliers_iqr(df.select_dtypes(include=['float64', 'int64']))

# Check the cleaned dataset's shape to see how many rows were removed
data_cleaned.shape
```

```python
sns.boxplot(data=data_cleaned)
plt.xticks(rotation=90)
```

```python
correlation_matrix = data_cleaned.corr()
plt.figure(figsize=(10, 8))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Batters Data')
plt.show()
```

```python
data_cleaned['Performance'] = (data_cleaned['BattingAvg'] +
    data_cleaned['TotalRuns'] +
    data_cleaned['StrikeRate'] 
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


X_batters = data_cleaned[['Matches', 'BattingInnings', 'TotalBallFaced', 'NotOuts', 
              'Bestscore']] 
y_batters = data_cleaned['Performance']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_batters, y_batters, test_size=0.2, random_state=42)

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
print(y_pred[0:5])
# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Calculate RMSE
rmse = mean_squared_error(y_test, y_pred, squared=False)

# Calculate accuracy as a percentage
accuracy_percentage = r2 * 100

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
print("Model Accuracy Percentage: {:.2f}%".format(accuracy_percentage))
```

```python
from sklearn.ensemble import RandomForestRegressor
# Initialize and train the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)  # 100 trees
rf_model.fit(X_train, y_train)

# Step 6: Make predictions on the test set
y1_pred = rf_model.predict(X_test)

# Step 7: Evaluate the model's performance
mse = mean_squared_error(y_test, y1_pred)
mae = mean_absolute_error(y_test, y1_pred)
r2 = r2_score(y_test, y1_pred)

# Calculate RMSE
rmse = mean_squared_error(y_test, y1_pred, squared=False)

# Calculate accuracy as a percentage using R-squared
accuracy_percentage = r2 * 100

# Print the results
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
print("Model Accuracy Percentage: {:.2f}%".format(accuracy_percentage))
```

```python
from sklearn.svm import SVR

# Define the SVR model (with default RBF kernel)
svr_model = SVR(kernel='linear')

# Train the model on the training data
svr_model.fit(X_train, y_train)

# Make predictions on the test set
y2_pred = svr_model.predict(X_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y2_pred)
mae = mean_absolute_error(y_test, y2_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y2_pred)
accuracy = r2 * 100  

# Print the performance metrics
print(f"Support Vector Regression (SVR) Performance:")
print(f"Mean Squared Error: {mse:.2f}")
print("Mean Absolute Error:", mae)
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R-squared: {r2:.4f}")
print(f"Model Accuracy Percentage: {accuracy:.2f}%")
```
